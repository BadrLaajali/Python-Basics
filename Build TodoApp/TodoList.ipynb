```python
todos = []

while True:
    user_action = input("Type add, show, edit or exit :")
    if user_action == 'add':
        todo = input("Enter a todo:")
        todos.append(todo.strip()) #strip remove empty space ex: 'add '
    elif user_action in {'show', 'display'}:  #avec in il est possible de dire si show ou display
        for index, item in enumerate(todos): #enumerate va nous permettre d'avoir l'index(position) dans la liste
            item = item.title() #first letter capitilzed
            #We add 1 to display numbers to user starting with 1 and not with 0
            message = f"{index + 1}- {item.capitalize()}" #Vous pouvez utiliser des expressions entre des accolades {} à l'intérieur de l'f-string pour inclure des variables
            print(message)
    elif user_action == 'edit':
        number = int(input("Enter the number for todo to edit:")) #int convert value to int
        number = number - 1 #just to have a number for user if 1 we will have 0 for the list
        new_todo = input("Enter the new todo:")
        todos[number] = new_todo
    elif user_action == 'complete':
        number = int(input("Enter the number for the todo to complete:")) 
        number = number - 1 
        todo_deleted = todos.pop(number)
        print("You completed this todo", todo_deleted.title())
    elif user_action == 'exit':
        break
    else: #on gere le cas ou le user n'a pas saisie une bonne valeur
        print("You enter a wrong command")
print("bye")
```

```python
import csv
import os

file_name = "todos.csv"

# Vérifiez si le fichier CSV existe, sinon créez-le
if not os.path.exists(file_name):
    with open(file_name, mode="w", newline="") as file:
        pass

while True:
    user_action = input("Type add, show, edit or exit :")
    #if user_action == 'add':
    #if 'add' in user_action:
    if user_action.startswith('add '):
        todo = user_action[4:].strip() #opération de découpage (slicing) de chaîne de caractères. Elle signifie que nous voulons extraire une sous-chaîne de user_action à partir du quatrième caractère (en commençant par 0-index).
        with open(file_name, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([todo])
            
    elif user_action in {'show', 'display'}:  
        todos = [] # Réinitialiser la liste vu qu'on l'utilise juste ici
        with open(file_name, mode="r") as file: #Ouvrir le fichier et alimenter la liste
            reader = csv.reader(file)
            # next(reader, None) # A utilise si on a une ligne en-tête
            for row in reader:
                if row and row[0] not in todos:  # Vérifie si la ligne n'est pas vide et si la tâche n'est pas déjà dans la liste todos
                    todos.append(row[0]) #row[0] extrait le contenu de la première colonne (ou la première cellule) de la ligne actuellement lue dans le fichier CSV, puis todos.append(row[0]) ajoute cette valeur à la liste todos.
        for index, item in enumerate(todos, start=1): #Lire la liste et afficher le contenu, commencer la liste par 1 au lieu de 0
            item = item.title() 
            message = f"{index}- {item.capitalize()}" 
            print(message)

    elif user_action.startswith('edit '):
        try:
            number = int(user_action[5:])
            number -= 1
            todos = []
            #Ici on va lire le fichier csv et stocker la liste des todos dans une liste todos en évitant d'écrire les lignes vides
            with open(file_name, mode="r") as file:
                reader = csv.reader(file)
                for row in reader:
                    if row and row[0]: #ici on test si la ligne n'est pas vide pour l'écrire dans la liste
                        todos.append(row[0])
            #Ici on va vérifier que le numéro saisie par l'utilisateur est valide, pas inférieur à 0 et ne depasse pas la longueur de la liste
            if 0 <= number < len(todos):
                new_todo = input("Enter the new todo:").strip()
                todos[number] = new_todo #Mettre à jour la todo dans la liste
                # Write the updated todos back to the CSV file
                with open(file_name, mode="w", newline="") as file:
                    writer = csv.writer(file)
                    for task in todos:
                        writer.writerow([task])
            else:
                print("You enter a wrong todo number")
        except ValueError:
            print("You enter a wrong command")
            continue #Go to the begening

    elif user_action.startswith('complete '):
        try:
            number = int(input("Enter the number for the todo to complete:")) 
            #Amélioration : Gérer le fait si on entre une valeur qui n'existe pas
            number = number - 1
            todo_founded = False 
            todos = []
            todo_deleted = []
            # Lisez le fichier CSV et stockez les tâches dans une liste en évitant les lignes à supprimer
            with open(file_name, mode="r") as file:
                reader = csv.reader(file)
                for i, row in enumerate(reader, start=1):
                    if i != number:  # Évitez d'écrire dans la liste la ligne à supprimer
                        todos.append(row[0]) #ici le row c'est la colonne, vu qu'on travail avec un fichier csv on a des colonne et ici on boucle sur la colonne 0
                    else:
                        todo_deleted.append(row[0])
                        todo_founded = True
            if todo_founded:
                # Réécrivez tout le contenu de la liste dans le fichier CSV
                with open(file_name, mode="w", newline="") as file:
                    writer = csv.writer(file)
                    for task in todos:
                        writer.writerow([task])
                print("You completed this todo", todo_deleted[0].title())
            else:
                print("This todo does not exist")
        except ValueError:
            print("You enter a wrong command")
            continue #Go to the begening


    elif user_action == 'test':
        print("A utiliser pour tester des fonctions")
    elif user_action == 'exit':
        break
    else: 
        print("You enter a wrong command")
print("bye")
```

```python
import csv
import os
import datetime

# Vérifiez si le fichier CSV existe, sinon créez-le
def check_file(file_name):
    if not os.path.exists(file_name):
        with open(file_name, mode="w", newline="") as file:
            pass

def add_todo(user_action, file_name):
    todo = user_action[4:].strip() 
    with open(file_name, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([todo])

def show_todo(file_name):
    todos = [] 
    with open(file_name, mode="r") as file: 
        reader = csv.reader(file)
        for row in reader:
            if row and row[0] not in todos:  
                todos.append(row[0])
    for index, item in enumerate(todos, start=1): 
        item = item.title() 
        message = f"{index}- {item.capitalize()}" 
        print(message)

def edit_todo(file_name, user_action):
    try:
        number = int(user_action[5:])
        number -= 1
        todos = []
        with open(file_name, mode="r") as file:
            reader = csv.reader(file)
            for row in reader:
                if row and row[0]: 
                    todos.append(row[0])
        if 0 <= number < len(todos):
            new_todo = input("Enter the new todo:").strip()
            todos[number] = new_todo 
            with open(file_name, mode="w", newline="") as file:
                writer = csv.writer(file)
                for task in todos:
                    writer.writerow([task])
        else:
            print("You enter a wrong todo number")
    except ValueError:
        print("You enter a wrong command")

def complete_todo(file_name):
    try:
        number = int(input("Enter the number for the todo to complete:")) 
        number = number - 1
        todo_founded = False 
        todos = []
        todo_deleted = []
        with open(file_name, mode="r") as file:
            reader = csv.reader(file)
            for i, row in enumerate(reader, start=1):
                if i != number:  
                    todos.append(row[0]) 
                else:
                    todo_deleted.append(row[0])
                    todo_founded = True
        if todo_founded:
            # Réécrivez tout le contenu de la liste dans le fichier CSV
            with open(file_name, mode="w", newline="") as file:
                writer = csv.writer(file)
                for task in todos:
                    writer.writerow([task])
            print("You completed this todo", todo_deleted[0].title())
        else:
            print("This todo does not exist")
    except ValueError:
        print("You enter a wrong command")


file_name = "todos.csv"
check_file(file_name)
# Obtenir la date et l'heure actuelles
now = datetime.datetime.now() # 2023-12-23 22:09:52.633851
# Si vous souhaitez formater la date et l'heure d'une manière particulière, vous pouvez utiliser strftime :
formatted_date = now.strftime("%Y-%m-%d") # 2023-12-23
formatted_time = now.strftime("%H:%M:%S") # 22:09:52
formatted_date_time = now.strftime("%Y-%m-%d %H:%M:%S")
# Afficher la date et l'heure au format par défaut
print("Date et heure actuelles : ", formatted_date_time)

while True:
    user_action = input("Type add and your todo, show, edit and the number of todo, complete and the number or exit :")
    if user_action.startswith('add '):
        add_todo(user_action, file_name)
    elif user_action in {'show', 'display'}:  
        show_todo(file_name)
    elif user_action.startswith('edit '):
        edit_todo(file_name, user_action)
    elif user_action.startswith('complete '):
        complete_todo(file_name)
    elif user_action == 'test':
        print("A utiliser pour tester des fonctions")
    elif user_action == 'exit':
        break
    else: 
        print("You enter a wrong command")
print("bye")
```

```python
import csv
import os

file_name = "./files/average.csv"

def check_file(file_name=file_name): #using default argument
    """Vérifiez si le fichier CSV existe, sinon créez-le""" #Voici comment déclarer un doc-string(help) pour une fonction
    if not os.path.exists(file_name):
        with open(file_name, mode="w", newline="") as file:
            pass

def get_list_value(file_name=file_name):
    list_value = []
    try:
        with open(file_name, mode="r") as file: 
            reader = csv.reader(file)
            next(reader) #Sauter la premiere ligne du fichier csv
            for row in reader:
                if row and row[0] not in list_value: #Vérifier si pas de ligne vide ET pas de doublons
                    try:
                        list_value.append(float(row[0])) #Convertir la valeur dans le fichier texte vers un float
                    except ValueError:
                        print(f"Ignoré : '{row[0]}' n'est pas convertible en float.")
                        continue
    except StopIteration:
        print("Votre fichier est vide")
    return list_value
     
def get_average(list_value):
    try:
        average = 0
        sum_value = sum(list_value)
        average = sum_value / len(list_value) #Average = Total / sur le nombre
    except ZeroDivisionError:
        print ("Division par Zero n'est pas autoriser")
    return average


check_file() #function declared with default argument so we dont need to pass it again

list_value = get_list_value("./files/average.csv") #In the function declaration we used default argument but we can also change it when we call the function if it's different
#print(list_value)

average_value = get_average(list_value)

#print(round(average_value))
print(round(average_value, 1)) #On va arrandir vers 1 valeur decimale
```
